```python
from BOM import ExtractBOM
noarlunga = ExtractBOM("sa", "adelaide", "Noarlunga")
mount_lofty = ExtractBOM("sa", "adelaide", "Mount Lofty")
temp = {
    "Noarlunga":noarlunga.extractByField("Temp", float)[0:162],
    "Mount Lofty":mount_lofty.extractByField("Temp", float)
}

print(len(temp["Noarlunga"]))
print(len(temp["Mount Lofty"]))
print(temp)
```

###### With this data, we can apply a number of statistical methods

```python
import matplotlib.pyplot as plt
x, = plt.plot(temp["Noarlunga"], label = "Noarlunga")
y, = plt.plot(temp["Mount Lofty"], label = "Mount Lofty")
plt.legend(handles = [x, y])
plt.title("Temperature")
plt.show()
plt.close()
```

###### Hypothesis: Higher wind speeds decrease temperature
First, let's test the correlation

```python
spd = {
    "Noarlunga":noarlunga.extractByField("Spd km/h", float)[0:162],
    "Mount Lofty":mount_lofty.extractByField("Spd km/h", float)
}
print(spd)
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
import numpy as np
from sklearn.preprocessing import normalize
print("Correlation (Noarlunga): {}".format(np.correlate(temp["Noarlunga"], spd["Noarlunga"])[0]))
print("Correlation (Mount Lofty): {}".format(np.correlate(temp["Mount Lofty"], spd["Mount Lofty"])[0]))
softmax = lambda X: [np.exp(x)/sum([np.exp(i) for i in X]) for x in X]
x, = plt.plot(normalize(temp["Noarlunga"]).reshape(-1, 1), label = "Temperature")
y, = plt.plot(normalize(spd["Noarlunga"]).reshape(-1, 1), label = "Speed")
plt.legend(handles = [x, y])
plt.title("Correlation Noarlunga")
plt.show()
plt.close()
x, = plt.plot(normalize(temp["Mount Lofty"]).reshape(-1, 1), label = "Temperature")
y, = plt.plot(normalize(spd["Mount Lofty"]).reshape(-1, 1), label = "Speed")
plt.legend(handles = [x, y])
plt.title("Correlation Mount Lofty")
plt.show()
plt.close()
```

###### Based on these results, it seems like wind speed is not a good explanation for the lower temperature on Mount Lofty

```python
x, = plt.plot(spd["Noarlunga"], label = "Noarlunga")
y, = plt.plot(spd["Mount Lofty"], label = "Mount Lofty")
plt.legend(handles = [x, y])
plt.title("Wind Speed in km/h")
plt.show()
plt.close()
```

###### It seems that my hypothesis is inadequate

```python
x, = plt.plot([a*b for a,b in zip(temp["Noarlunga"], spd["Noarlunga"])], label = "Noarlunga")
y, = plt.plot([a*b for a,b in zip(temp["Mount Lofty"], spd["Mount Lofty"])], label = "Mount Lofty")
plt.legend(handles = [x, y])
plt.title("Wind Speed Times Temperature")
plt.show()
plt.close()
```
